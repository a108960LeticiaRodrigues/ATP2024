### TPC6: Aplicação para gestão de alunos

Considere que o modelo do aluno e da turma têm a seguinte estrutura:

`aluno = (nome, id, [notaTPC, notaProj, notaTeste])`

`turma = [aluno]`

* Cria uma aplicação que coloca no monitor o seguinte menu de operações:
    - 1: Criar uma turma;
    - 2: Inserir um aluno na turma;
    - 3: Listar a turma;
    - 4: Consultar um aluno por id;
    - 8: Guardar a turma em ficheiro;
    - 9: Carregar uma turma dum ficheiro;
    - 0: Sair da aplicação
* No fim de executar a operação selecionada, a aplicação deverá colocar novamente o menu e pedir ao utilizador a opção para continuar;
* Utiliza a tua aplicação para criar uma turma com 5 alunos.

```python
def menu():
    print("-----------MENU-------------")
    print("1: Criar uma turma")
    print("2: Inserir um aluno na turma")
    print("3: Listar a turma")
    print("4: Consultar um aluno por id")
    print("5: Guardar a turma em ficheiro")
    print("6: Carregar uma turma dum ficheiro")
    print("0: Sair da aplicação")

def criar_turma():
    return []


def Inserir_Aluno(turma):
        nome = input("Qual é o nome do aluno?")
        id = input("Qual é o id do aluno? (aXXXXXX)")

        id_existe = False
        notas_validas = True

        for aluno in turma:
            if aluno[1] == id:
                print("Erro: O ID já existe. A inserção do aluno foi cancelada.")
                id_existe = True
        if not id_existe:
            notaTPC = float(input("Qual é a nota do TPC do aluno? (de 0 a 20)"))
            if notaTPC > 20 or notaTPC < 0:
                notas_validas = False 
            notaProj = float(input("Qual é a nota do projeto do aluno? (de 0 a 20)"))
            if notaProj > 20 or notaProj < 0:
                notas_validas = False 
            notaTeste = float(input("Qual é a nota do teste do aluno? (de 0 a 20)"))
            if notaTeste > 20 or notaTeste < 0:
                notas_validas = False 

        if notas_validas and not id_existe: 
            notas = [notaTPC, notaProj, notaTeste]
            aluno = (nome,id,notas)
            turma.append(aluno)
            print(turma)
        elif not notas_validas: 
            print(f"Aluno inválido pois notas não são validas (só entre 0 a 20)")

def Consultar_Aluno(ID_aluno): 
    for aluno in turma:
        if aluno[1] == ID_aluno:
            print(f"O aluno foi encontrado: {aluno}")
        else:
            print("ID não existe.")

def Escrever_Turma(ficheiro_turma):
    file = open(ficheiro_turma,"w")
    for aluno in turma:
        nome, id, notas = aluno
        file.write(str(nome) + "|" + str(id) + "|" + str(notas[0]) + "|" + str(notas[1]) + "|" + str(notas[2]) + "\n" )
    file.close()
    print(f"Turma guardada em {ficheiro_turma} com sucesso!")

def Carregar_Turma(ficheiro_turma):
    turma = []
    file = open(ficheiro_turma,"r")
    for line in file:
        elementos = line.split('|') 
        nome = elementos[0]
        id = elementos[1]
        notas = [float(elementos[2]), float(elementos[3]), float(elementos[4])]
        aluno = (nome,id,notas)
        turma.append(aluno)
    file.close()
    return turma

cond = True

while cond:
    menu()
    opção = int(input("Qual opção deseja executar?"))

    if opção == 1:
        turma = criar_turma()
        print("Turma criada com sucesso!")

    elif opção == 2:
        Inserir_Aluno(turma)

    elif opção == 3:
        if turma != []:
            for aluno in turma:
                print(aluno)
        else:
            print(turma)

    elif opção == 4:
        ID_aluno = int(input("Qual é o Id do aluno que deseja procurar? "))
        Consultar_Aluno(ID_aluno)
        
    elif opção == 5:
        ficheiro_turma = input("Introduza o nome do ficheiro onde deseja guardar a turma: ")
        Escrever_Turma(ficheiro_turma)

    elif opção == 6:
        print(Carregar_Turma(ficheiro_turma))
    
    elif opção == 0:
        print("Volte sempre!")
        cond = False

    elif opção < 0 or opção > 6:
        print("Escolha inválida, introduza um número entre 0 e 6.")
```
